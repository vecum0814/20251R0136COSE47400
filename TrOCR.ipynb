## Box 기준으로 이미지 크롭하여 저장하기

```python
# box 위치 테스트
from PIL import Image, ImageDraw
import json
import os

def visualize_bbox(json_path, image_path, save_path="visualized.jpg"):
    with open(json_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    image = Image.open(image_path).convert("RGB")
    draw = ImageDraw.Draw(image)

    for ann in data["annotations"]:
        x, y, w, h = ann["bbox"]
        x1, y1 = int(x), int(y)
        x2, y2 = int(x + w), int(y + h)

        draw.rectangle([x1, y1, x2, y2], outline="red", width=2)
        draw.text((x1, y1), ann["text"], fill="blue")


    # 저장해서 확인
    image.save(save_path)
    print(f"시각화 결과 저장 완료: {save_path}")
```

```python
#실제 box 범위대로 crop하기
import os
import json
from PIL import Image, ImageOps

def crop_boxes_per_sample(json_path, image_path, output_dir_name, output_root):
    # 저장 경로 설정
    save_root = os.path.join(output_root, output_dir_name)
    image_save_dir = os.path.join(save_root, "images")
    label_save_dir = os.path.join(save_root, "labels")
    os.makedirs(image_save_dir, exist_ok=True)
    os.makedirs(label_save_dir, exist_ok=True)

    # JSON 로드
    with open(json_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    # 이미지 열고 회전 보정
    image = Image.open(image_path)
    image = ImageOps.exif_transpose(image)  # ← 회전 보정 추가
    image = image.convert("RGB")
    width, height = image.size

    saved = 0
    for idx, ann in enumerate(data["annotations"]):
        x, y, w, h = ann["bbox"]
        text = ann["text"]

        # 좌표 정리 및 clipping
        x1 = int(max(0, x))
        y1 = int(max(0, y))
        x2 = int(min(x + w, width))
        y2 = int(min(y + h, height))

        if x2 <= x1 or y2 <= y1:
            print(f"무시된 잘못된 bbox: {(x1, y1, x2, y2)}")
            continue

        try:
            crop = image.crop((x1, y1, x2, y2))
            img_name = f"{output_dir_name}_{idx+1:03}.jpg"
            txt_name = f"{output_dir_name}_{idx+1:03}.txt"

            crop.save(os.path.join(image_save_dir, img_name))
            with open(os.path.join(label_save_dir, txt_name), "w", encoding="utf-8") as f:
                f.write(text)

            saved += 1

        except Exception as e:
            print(f"{output_dir_name}_{idx+1:03} 저장 실패: {e}")

    print(f"{output_dir_name}: {saved}개 bbox 저장 완료")
```

```python
# 전체 데이터에 대하여 crop 및 저장
import os

def batch_crop_boxes(json_dir, image_dir, output_root):
    json_files = sorted([f for f in os.listdir(json_dir) if f.endswith(".json")])
    possible_exts = [".jpg", ".JPG", ".jpeg", ".png"]

    for json_file in json_files:
        base_name = os.path.splitext(json_file)[0]  # ex: 책표지_총류_000001
        json_path = os.path.join(json_dir, json_file)

        # 이미지 경로 찾기 (.jpg, .JPG 등 자동 탐색)
        image_path = None
        for ext in possible_exts:
            candidate = os.path.join(image_dir, base_name + ext)
            if os.path.exists(candidate):
                image_path = candidate
                break

        if image_path is None:
            print(f"❌ 이미지 없음: {base_name}.[jpg|JPG|...]")
            continue

        # output_dir_name = 000001
        output_dir_name = base_name.split("_")[-1]

        crop_boxes_per_sample(
            json_path=json_path,
            image_path=image_path,
            output_dir_name=output_dir_name,
            output_root=output_root
        )
```

```python
# 실행
json_dir = "AIHub_OCR_sample/라벨링데이터/[라벨]Training/2.책표지/01.총류"
image_dir = "AIHub_OCR_sample/원천데이터/[원천]Training_책표지1/01.총류"
output_root = "cropped_all"

batch_crop_boxes(json_dir, image_dir, output_root)
```

# AI Hub OCR Dataset 준비 (LQ)

```python
import os
from PIL import Image, ImageOps

def compress_aihub_images(
    hq_dir,         # 원본 이미지 디렉토리
    out_dir,        # 압축 이미지 저장 디렉토리
    scale=0.25,     # 해상도 축소 비율
    quality=90      # JPEG 압축 품질
):
    os.makedirs(out_dir, exist_ok=True)

    valid_ext = ('.png', '.jpg', '.jpeg', '.bmp')
    file_list = sorted([f for f in os.listdir(hq_dir) if f.lower().endswith(valid_ext)])

    for fname in file_list:
        hq_path = os.path.join(hq_dir, fname)
        try:
            # 이미지 로드 + EXIF 회전 보정
            img_raw = Image.open(hq_path)
            img = ImageOps.exif_transpose(img_raw).convert('RGB')

            w, h = img.size
            new_size = (int(w * scale), int(h * scale))
            img_resized = img.resize(new_size, Image.LANCZOS)

            # 파일명 그대로 저장
            out_path = os.path.join(out_dir, fname)
            img_resized.save(out_path, "JPEG", quality=quality)

        except Exception as e:
            print(f"{fname} 처리 실패: {e}")

    print(f"\n압축 완료: {len(file_list)}개 → {out_dir}")
```

```python
#실제 box 범위대로 crop하기
import os
import json
from PIL import Image, ImageOps

def crop_boxes_per_sample(json_path, image_path, output_dir_name, output_root):
    # 저장 경로 설정
    save_root = os.path.join(output_root, output_dir_name)
    image_save_dir = os.path.join(save_root, "images")
    label_save_dir = os.path.join(save_root, "labels")
    os.makedirs(image_save_dir, exist_ok=True)
    os.makedirs(label_save_dir, exist_ok=True)

    # JSON 로드
    with open(json_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    # 이미지 열고 회전 보정
    image = Image.open(image_path)
    image = ImageOps.exif_transpose(image)  # ← 회전 보정 추가
    image = image.convert("RGB")
    width, height = image.size

    saved = 0
    for idx, ann in enumerate(data["annotations"]):
        x, y, w, h = ann["bbox"]
        text = ann["text"]
        
        # 스케일 적용
        x1 = x * scale
        y1 = y * scale
        x2 = (x + w) * scale
        y2 = (y + h) * scale
        
        # 좌표 정리 및 clipping
        x1 = int(max(0, x))
        y1 = int(max(0, y))
        x2 = int(min(x + w, width))
        y2 = int(min(y + h, height))

        if x2 <= x1 or y2 <= y1:
            print(f"무시된 잘못된 bbox: {(x1, y1, x2, y2)}")
            continue

        try:
            crop = image.crop((x1, y1, x2, y2))
            img_name = f"{output_dir_name}_{idx+1:03}.jpg"
            txt_name = f"{output_dir_name}_{idx+1:03}.txt"

            crop.save(os.path.join(image_save_dir, img_name))
            with open(os.path.join(label_save_dir, txt_name), "w", encoding="utf-8") as f:
                f.write(text)

            saved += 1

        except Exception as e:
            print(f"{output_dir_name}_{idx+1:03} 저장 실패: {e}")

    print(f"{output_dir_name}: {saved}개 bbox 저장 완료")
```

```python
#box영역 기준 각 텍스트 단위로 이미지 크롭하여 저장(wordbox단위)
import os
import json
from PIL import Image

def crop_boxes_with_labels(json_path, image_path, output_root):
    # json 로드
    with open(json_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    image = Image.open(image_path).convert("RGB")
    base_name = os.path.splitext(os.path.basename(image_path))[0]

    # 저장 폴더 생성
    os.makedirs(os.path.join(output_root, "images"), exist_ok=True)
    os.makedirs(os.path.join(output_root, "labels"), exist_ok=True)

    saved = 0
    for idx, word in enumerate(data["text"]["word"]):
        box = word["wordbox"]  # [x1, y1, x2, y2]
        value = word["value"]

        cropped = image.crop(box)

        # 저장 경로 지정
        crop_name = f"{base_name}_{idx:03}.jpg"
        label_name = f"{base_name}_{idx:03}.txt"

        cropped.save(os.path.join(output_root, "images", crop_name))
        with open(os.path.join(output_root, "labels", label_name), "w", encoding="utf-8") as f:
            f.write(value)

        saved += 1

    print(f"총 {saved}개의 crop 저장 완료: {output_root}")
```

```python
import os
import pandas as pd
from tqdm import tqdm
from jiwer import wer, cer
from PIL import Image

def evaluate_ocr_on_cropped_set(
    cropped_root,
    predict_fn,
    output_csv="ocr_crop_eval.csv"
):
    image_dir = os.path.join(cropped_root, "images")
    label_dir = os.path.join(cropped_root, "labels")

    image_files = sorted([
        f for f in os.listdir(image_dir) if f.endswith((".jpg", ".png"))
    ])

    results = []

    for fname in tqdm(image_files):
        img_path = os.path.join(image_dir, fname)
        label_path = os.path.join(label_dir, fname.rsplit(".", 1)[0] + ".txt")

        if not os.path.exists(label_path):
            print(f"라벨 누락: {label_path}")
            continue

        # 정답 불러오기
        with open(label_path, "r", encoding="utf-8") as f:
            gt = f.read().strip()

        # 예측 수행
        pred = predict_fn(img_path)

        # 평가 저장
        results.append({
            "file": fname,
            "GT": gt,
            "Pred": pred,
            "WER": round(wer(gt, pred), 4),
            "CER": round(cer(gt, pred), 4),
            "Top1Acc": int(gt == pred)
        })

    # 결과 저장
    df = pd.DataFrame(results)
    df.to_csv(output_csv, index=False, encoding="utf-8-sig")
    print(f"평가 완료! 결과 저장됨 → {output_csv}")
    return df
```

## GT 추출 함수

### 목적 데이터용

```python
import json

def extract_gt_text(json_path):
    with open(json_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
    annotations = sorted(data["annotations"], key=lambda x: x["id"])
    gt_text = " ".join([ann["text"] for ann in annotations])
    return gt_text
```

```python
import os
import pandas as pd
from jiwer import wer, cer
from tqdm import tqdm
from PIL import Image
from transformers import TrOCRProcessor, VisionEncoderDecoderModel
import unicodedata

# 텍스트 정규화 함수
def normalize(text):
    text = unicodedata.normalize("NFKC", text)  # 유니코드 정규화
    return ''.join(text.strip().split())        # 모든 공백 제거

# 모델 로딩
processor = TrOCRProcessor.from_pretrained("ddobokki/ko-trocr")
model = VisionEncoderDecoderModel.from_pretrained("ddobokki/ko-trocr")
model.eval()

# OCR 함수
def predict_ko_trocr(image_path):
    image = Image.open(image_path).convert("RGB")
    pixel_values = processor(images=image, return_tensors="pt").pixel_values
    generated_ids = model.generate(pixel_values, max_new_tokens=64)
    pred_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
    return pred_text.strip()

# 루프 시작
root_dir = "cropped_all"
sample_dirs = sorted([d for d in os.listdir(root_dir) if os.path.isdir(os.path.join(root_dir, d))])
results = []

for sample in tqdm(sample_dirs):
    image_dir = os.path.join(root_dir, sample, "images")
    label_dir = os.path.join(root_dir, sample, "labels")

    image_files = sorted([f for f in os.listdir(image_dir) if f.endswith(".jpg")])
    gt_list = []
    pred_list = []

    for fname in image_files:
        img_path = os.path.join(image_dir, fname)
        txt_path = os.path.join(label_dir, fname.replace(".jpg", ".txt"))

        if not os.path.exists(txt_path):
            continue

        with open(txt_path, "r", encoding="utf-8") as f:
            gt_list.append(f.read().strip())

        pred = predict_ko_trocr(img_path)
        pred_list.append(pred)

    # 원본
    gt_raw = " ".join(gt_list)
    pred_raw = " ".join(pred_list)

    # 정규화
    gt_norm = normalize(gt_raw)
    pred_norm = normalize(pred_raw)

    # 평가 지표
    results.append({
        "sample": sample,
        "GT": gt_raw,
        "Pred": pred_raw,
        "WER": round(wer(gt_norm, pred_norm), 4),
        "CER": round(cer(gt_norm, pred_norm), 4),
        "Top1Acc": int(gt_norm == pred_norm)
    })

# 저장
df = pd.DataFrame(results)
output_path = "trocr_concat_eval_results.csv"
df.to_csv(output_path, index=False, encoding="utf-8-sig")

print(f"\n저장 완료: {output_path} (총 {len(df)}개 샘플)")
```

```python
import json

def extract_gt_text(json_path):
    with open(json_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
    annotations = sorted(data["annotations"], key=lambda x: x["id"])
    gt_text = " ".join([ann["text"] for ann in annotations])
    return gt_text
```

```python
import os
import pandas as pd
from jiwer import wer, cer
from tqdm import tqdm
from PIL import Image, ImageOps
from transformers import TrOCRProcessor, VisionEncoderDecoderModel
import unicodedata

# 텍스트 정규화 함수
def normalize(text):
    text = unicodedata.normalize("NFKC", text)
    return ''.join(text.strip().split())

# 모델 로딩
processor = TrOCRProcessor.from_pretrained("ddobokki/ko-trocr")
model = VisionEncoderDecoderModel.from_pretrained("ddobokki/ko-trocr")
model.eval()

# OCR 함수 (예외 처리 추가)
def predict_ko_trocr(image_path):
    try:
        image = Image.open(image_path)
        image = ImageOps.exif_transpose(image).convert("RGB")  # 회전 보정
        pixel_values = processor(images=image, return_tensors="pt").pixel_values
        generated_ids = model.generate(pixel_values, max_new_tokens=64)
        pred_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
        return pred_text.strip()
    except Exception as e:
        print(f"[ERROR] {image_path} 처리 중 오류 발생: {e}")
        return ""

# 루프 시작 (압축 이미지 기반)
root_dir = "cropped_all_Low"  # 압축된 crop 데이터 폴더
sample_dirs = sorted([d for d in os.listdir(root_dir) if os.path.isdir(os.path.join(root_dir, d))])
results = []

for sample in tqdm(sample_dirs):
    image_dir = os.path.join(root_dir, sample, "images")
    label_dir = os.path.join(root_dir, sample, "labels")

    image_files = sorted([
        f for f in os.listdir(image_dir)
        if f.lower().endswith(".jpg")  # .JPG 포함
    ])

    gt_list = []
    pred_list = []

    for fname in image_files:
        img_path = os.path.join(image_dir, fname)
        txt_path = os.path.join(label_dir, fname.rsplit(".", 1)[0] + ".txt")

        if not os.path.exists(txt_path):
            print(f"[WARNING] 정답 라벨 없음: {txt_path}")
            continue

        with open(txt_path, "r", encoding="utf-8") as f:
            gt_list.append(f.read().strip())

        pred = predict_ko_trocr(img_path)
        pred_list.append(pred)

    # 원본 & 정규화
    gt_raw = " ".join(gt_list)
    pred_raw = " ".join(pred_list)
    gt_norm = normalize(gt_raw)
    pred_norm = normalize(pred_raw)

    # 평가 지표
    results.append({
        "sample": sample,
        "GT": gt_raw,
        "Pred": pred_raw,
        "WER": round(wer(gt_norm, pred_norm), 4),
        "CER": round(cer(gt_norm, pred_norm), 4),
        "Top1Acc": int(gt_norm == pred_norm)
    })

# 저장
df = pd.DataFrame(results)
output_path = "trocr_concat_eval_results_LQ.csv"
df.to_csv(output_path, index=False, encoding="utf-8-sig")

print(f"\n저장 완료: {output_path} (총 {len(df)}개 샘플)")
```

# HQ vs. LQ 성능 비교

```python
import pandas as pd

# 결과 파일 경로
hq_path = "trocr_concat_eval_results.csv"
lq_path = "trocr_concat_eval_results_LQ.csv"

# 결과 불러오기
hq_df = pd.read_csv(hq_path)
lq_df = pd.read_csv(lq_path)

# sample 기준 정렬
hq_df = hq_df.sort_values("sample").reset_index(drop=True)
lq_df = lq_df.sort_values("sample").reset_index(drop=True)

# 'x'가 GT에 포함된 샘플 제거
valid_mask = ~hq_df["GT"].str.contains("x")
hq_df = hq_df[valid_mask].reset_index(drop=True)
lq_df = lq_df[valid_mask].reset_index(drop=True)

# 비교 결과 DataFrame 구성
comparison = hq_df[["sample", "GT", "Pred"]].copy()
comparison = comparison.rename(columns={"Pred": "Pred_HQ"})
comparison["Pred_LQ"] = lq_df["Pred"]
comparison["WER_HQ"] = hq_df["WER"]
comparison["WER_LQ"] = lq_df["WER"]
comparison["CER_HQ"] = hq_df["CER"]
comparison["CER_LQ"] = lq_df["CER"]
comparison["Top1_HQ"] = hq_df["Top1Acc"]
comparison["Top1_LQ"] = lq_df["Top1Acc"]

# 비교 함수
def compare(hq_val, lq_val, higher_is_better=False):
    if hq_val == lq_val:
        return "같음"
    if higher_is_better:
        return "HQ 우세" if hq_val > lq_val else "LQ 우세"
    else:
        return "HQ 우세" if hq_val < lq_val else "LQ 우세"

# 각 메트릭별 비교 결과
comparison["WER_우위"] = comparison.apply(lambda row: compare(row["WER_HQ"], row["WER_LQ"]), axis=1)
comparison["CER_우위"] = comparison.apply(lambda row: compare(row["CER_HQ"], row["CER_LQ"]), axis=1)
comparison["Top1_우위"] = comparison.apply(lambda row: compare(row["Top1_HQ"], row["Top1_LQ"], higher_is_better=True), axis=1)

# 저장
output_path = "trocr_HQ_vs_LQ_comparison_filtered.csv"
comparison.to_csv(output_path, index=False, encoding="utf-8-sig")
print(f"비교 결과 저장 완료: {output_path}")

# -------------------------------------
# 요약 표 생성 (print용)
# -------------------------------------
hq_mean_wer = hq_df["WER"].mean()
lq_mean_wer = lq_df["WER"].mean()
hq_mean_cer = hq_df["CER"].mean()
lq_mean_cer = lq_df["CER"].mean()
hq_top1_acc = hq_df["Top1Acc"].mean()
lq_top1_acc = lq_df["Top1Acc"].mean()

print("\n아래는 x가 포함된 GT 샘플을 제외한 후, HQ vs LQ OCR 성능 비교 결과입니다:\n")
print(f"{'Metric':<15}{'HQ':>10}{'LQ':>10}")
print("-" * 35)
print(f"{'WER (mean)':<15}{hq_mean_wer:>10.4f}{lq_mean_wer:>10.4f}")
print(f"{'CER (mean)':<15}{hq_mean_cer:>10.4f}{lq_mean_cer:>10.4f}")
print(f"{'Top1 Accuracy':<15}{hq_top1_acc:>10.4f}{lq_top1_acc:>10.4f}")
```
